### Building the Classifier

We are now going to build the baseline classifier that you will use throughout this homework.

```python
# Data normalization
x_train = x_train/255
x_test = x_test/255
```

```python
inputs = Input(shape=(28, 28, 1))
net = Conv2D(32, kernel_size=(3, 3), activation="relu", padding='same')(inputs)
net = Flatten()(net)
net = Dense(128)(net)
outputs = Dense(10, activation="softmax")(net)

model = Model(inputs, outputs)
model.summary()
```

```python
epochs = 50
batch_size = 64

model.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
history = model.fit(x_train, y_train, batch_size=batch_size, epochs=epochs, validation_data=(x_test, y_test))
```

```python
def plot_history(history):
    h = history.history
    epochs = range(len(h['loss']))

    plt.subplot(121), plt.plot(epochs, h['loss'], '.-', epochs, h['val_loss'], '.-')
    plt.grid(True), plt.xlabel('epochs'), plt.ylabel('loss')
    plt.legend(['Train', 'Validation'])
    plt.subplot(122), plt.plot(epochs, h['accuracy'], '.-',
                               epochs, h['val_accuracy'], '.-')
    plt.grid(True), plt.xlabel('epochs'), plt.ylabel('Accuracy')
    plt.legend(['Train', 'Validation'])
        
    print('Train Acc     ', h['accuracy'][-1])
    print('Validation Acc', h['val_accuracy'][-1])
    
plot_history(history)
```

As you can see, the classifier suffers from massive overfitting. The validation accuracy is around 88% while the training accuracy is close to 1.

### Combat the Overfitting!

There are to objective you shall fulfill in order to successfully complete this homework:

* Your network (with all the regularizations applied) shall **not** be larger than the baseline

```python
# Modify the baseline classifier in order to reduce the overfitting and make the performance more robust

from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Dropout, Flatten, Dense, BatchNormalization, Activation
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping

inputs = Input(shape=(28, 28, 1))

net = Conv2D(64, kernel_size=(5, 5), padding='same')(inputs)
net = BatchNormalization()(net) #  Batch Normalization
net = Activation('relu')(net)  #   
net = MaxPooling2D(pool_size=(2, 2))(net)
net = Dropout(0.2)(net)
net = Flatten()(net)
net = Dense(128)(net)
net = BatchNormalization()(net)  
net = Activation('relu')(net) 
net = Dropout(0.3)(net)
outputs = Dense(10, activation="softmax")(net)

model = Model(inputs, outputs)
model.summary()
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
data_augmentation = ImageDataGenerator(
    rotation_range=10,       #      10 
    zoom_range=0.1,         #  
    horizontal_flip=True,   #   
    fill_mode='nearest'     #    ,   '  
)
x_train = np.expand_dims(x_train, -1) 
x_test = np.expand_dims(x_test, -1) 
```

```python
from tensorflow.keras.callbacks import Callback

class CustomEarlyStopping(Callback):
    def __init__(self, target_val_acc=0.915):
        super(CustomEarlyStopping, self).__init__()
        #      
        self.target_val_acc = target_val_acc

    def on_epoch_end(self, epoch, logs=None):
        #      
        val_acc = logs.get('val_accuracy')
        if val_acc is not None:
            # ,        
            if val_acc > self.target_val_acc:
                print(f"\n    {val_acc:.4f}    {self.target_val_acc}.  .")
                self.model.stop_training = True

custom_early_stopping = CustomEarlyStopping(target_val_acc=0.915)
```

```python
from tensorflow.keras.optimizers import Adam
learning_rate = 0.0015
optimizer = Adam(learning_rate=learning_rate)
```

```python
steps_per_epoch = x_train.shape[0] // batch_size
```

```python
# Train the network
epochs = 50
batch_size = 128
model.compile(loss="sparse_categorical_crossentropy", optimizer=optimizer, metrics=["accuracy"])
history = model.fit(
    data_augmentation.flow(x_train, y_train, batch_size=batch_size),
    validation_data=(x_test, y_test),
    epochs=epochs,
    steps_per_epoch=steps_per_epoch,
    callbacks=[custom_early_stopping]
)
```

```python
# Show the results
plot_history(history)
```
